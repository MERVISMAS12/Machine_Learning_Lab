```python
import pandas as pd
iris = pd.read_csv('/content/Iris_Dataset.csv')
iris.head()
```

```python
from sklearn.cluster import KMeans
```

```python
x = iris.drop('variety',axis=1)
y = iris['variety']
```

```python
km = KMeans(n_clusters=3)
km.fit(x)
```

```python
pred = km.fit_predict(x)
```

```python
pred
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size if needed

# Scatter plot of sepal length vs sepal width, colored by cluster
plt.scatter(x['sepal.length'], x['sepal.width'], c=pred, cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('KMeans Clustering of Iris Dataset')

# Add cluster centers to the plot
centers = km.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=200, label='Cluster Centers')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(x['petal.length'], x['petal.width'], c=pred, cmap='viridis')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('KMeans Clustering of Iris Dataset')
centers = km.cluster_centers_
plt.scatter(centers[:, 2], centers[:, 3], c='red', marker='x', s=200, label='Cluster Centers')
```
